```python
#replacement method
#outlier replacement with median
#1.find Q1 and Q3
q1=df.bmi.quantile(0.25)
q3=df.bmi.quantile(0.75)
print(q1)
print(q3)
```

```python
#IQR =q3-q1
iqr=q3-q1
print(iqr)
```

```python
#upper limit
upper_limit=q3+1.5*iqr
upper_limit
```

```python
#lower limit
lower_limit=q1-1.5*iqr
lower_limit
```

```python
#median
df.median()
```

```python
#replacing outlier with medain
df['bmi']=np.where(df['bmi']>upper_limit,30,df['bmi'])
```

```python
#checking
sns.boxplot(df.bmi)
```

```python
#done with outlier checking
```
